```python
import pandas as pd
```

```python
data1=pd.read_csv('abc_final_Daily.csv',index_col=0)
```

```python
data1
```

```python
data2=pd.read_csv('abc_final_hour.csv',index_col=0)
```

```python
data2
```

```python
data3=pd.read_csv('abc_final_min.csv',index_col=0)
```

```python
data3
```

```python
last=pd.merge(data1,data2,how='inner',left_on=['Id',],right_on=['Id'])
```

```python
last
```

```python
last.columns
```

```python
result = pd.concat([last, data3], ignore_index=True)
```

```python
result
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.violinplot(x='ActivityDay', y='Calories_x', data=result)
plt.title('Violin Plot of NumericColumn by CategoryColumn')
plt.xlabel('activity day')
plt.ylabel('calories')
plt.show()
```

```python
sns.countplot(x='ActivityDay', data=result)
plt.title('Count Plot of CategoryColumn')
plt.xlabel('CategoryColumn')
plt.ylabel('Count')
plt.show()
```

```python
correlation_matrix = result.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.boxplot(x=result['SedentaryMinutes_x'])
plt.title('Box Plot of NumericColumn')
plt.xlabel('NumericColumn')
plt.show()
```
