```python
import pandas as pd
```

```python
data1=pd.read_csv('cleaned_hourly-steps.csv',index_col=0)
```

```python
data1
```

```python
data2=pd.read_csv('cleaned_Hourly_calories.csv',index_col=0)
```

```python
data2
```

```python
data3=pd.read_csv('cleaned_Hourly_intensities.csv',index_col=0)
```

```python
data3
```

```python
data4_final=pd.merge(data1,data2,how='inner',left_on=['Id', 'date', 'time'],right_on=['Id', 'date', 'time'])
```

```python
data4_final
```

```python
data4_final_Last=pd.merge(data4_final,data3,how='inner',left_on=['Id', 'date', 'time'],right_on=['Id', 'date', 'time'])
```

```python
data4_final_Last
```
